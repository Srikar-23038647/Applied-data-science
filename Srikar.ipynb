```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

#read Srikar data set
df=pd.read_csv('Srikar.csv')

#describe the data set
df.describe()

#info of data set
df.info()
```

```python
#line graph of Weight and Size of the Apples

def graph_of_Weight():
    df_size=df['Size']

    plt.plot(df_size , df['Weight'] , label='Size & Weight')
    plt.xlabel('Size' ,color='red')
    plt.ylabel('weight' ,color='blue')
    plt.legend(loc="upper left")
    plt.title('Line graph of Apple data Set', color='blue')
    plt.show()
    
graph_of_Weight()
```

```python
#Scatter graph of id and Juiciness.
df_id=df['A_id']
plt.scatter(df_id , df['Juiciness'],label='Juiciness & ID' ,marker=".")

plt.xlabel('A_id',color="red")
plt.ylabel('Juiciness',color="blue")
plt.legend(loc="upper left")
plt.title('Scatter graph of Apple data Set', color='blue')
plt.show()
```

```python
# Box Plot for Each Column

plt.figure(dpi=125)

# Assuming df is your DataFrame
df.boxplot()

plt.title('Box Plot for Each Column' ,color="Green")
plt.ylabel('Values' ,color="blue")
plt.show()
```

```python
#create a histogram for a each column of DataFrame
plt.figure(figsize=(8, 4))
df.hist()
```

```python
#create a PIE CHART for Juiciness

data_quality = df.groupby('Quality').size()

plt.title("PIE CHART")
    
plt.pie(data_quality, labels=data_quality.index, autopct='%1.1f%%')

plt.show()
```
